```python
import csv
import unicodecsv as unicsv

# find the personid from the name
def find_personid(name):
    with open('db2018imdb/merge_people.csv', newline='', encoding='utf-8') as peoCSV:
        people = csv.reader(peoCSV, delimiter=',', quotechar='"')
        for id, fullname in people:
            if fullname == name:
                return id
        return 0

# find the last personid in [merge_people.csv], in order to add new people without a profession
row_count = len((open('db2018imdb/merge_people.csv', newline='', encoding='utf-8')).readlines())
id = row_count+1

# used to print where is the process, to be deleted afterward
term = 1
bioterm = 1
infoterm = 1

# create [merge_biographies.csv], [merge_bioinfos.csv]
# complete [merge_people.csv]
def add_bioinfos():
    with open('db2018imdb/biographies.csv', newline='', encoding='utf-8') as bioCSV, \
    open('db2018imdb/merge_people.csv', mode='ab') as peopleCSV, \
    open('db2018imdb/merge_biographies.csv', mode='wb') as dstbio,\
    open('db2018imdb/merge_bioinfo.csv', mode='wb') as dstinfo:
      
#[Bioinfos.csv] should look like this:
# personid INTEGER UNIQUE NOT NULL,
# realname VARCHAR(100),
#   nickname VARCHAR(100),
#   trademark VARCHAR(100),
#   birth VARCHAR(100),
#   death VARCHAR(100),
#   salary INTEGER,
#   whereAreTheyNow VARCHAR(100),
#   height VARCHAR(50),
#   spouse VARCHAR(100),
#   biographicalBooks VARCHAR(1000),
#   trivia VARCHAR(500),
#   personalQuote VARCHAR(500),  

# [Biographie.csv] given like this:
# Name, RealName, Nickname, DateAndPlaceOfBirth, Height, [0-4]
# Biography,Biographer, DateAndCauseOfDeath, Spouse, Trivia, [5-9]
# BiographicalBooks, PersonalQuotes, Salary, Trademark, WhereAreTheyNow [10-14]

        fieldnames = ['personid', 'realname', 'nickname', 'trademark', 'birth', 'death','salary', \
                      'whereAreTheyNow', 'height', 'spouse', 'biographicalBooks', 'trivia','personalQuote']
        bioinfos = unicsv.writer(dstinfo, delimiter=",", quotechar='"', encoding='utf-8')
        people = unicsv.writer(peopleCSV, delimiter=",", quotechar='"', encoding='utf-8')
        biograhies = unicsv.writer(dstbio, delimiter=",", quotechar='"', encoding='utf-8')
        
        bios = csv.reader(bioCSV, delimiter=',', quotechar='"')
                
        global id
        global bioterm
        global infoterm
        global term
        
        for n, rl, nm, birth, h, phy, pher, death, sp, tri, books, quotes, sal, trad, where in bios:
            
            print("Term ----------------------------- ", term)
            term = term + 1
            
            personid = find_personid(n)
            
            if personid == 0:  #the person is not in [merge_people.csv] yet
                people.writerow([id, n])
                
                if not (rl=='' and nm=='' and birth=='' and h=='' and\
                       death=='' and sp=='' and tri=='' and books=='' and \
                        quotes=='' and sal=='' and  trad=='' and where==''): # if there is bioinfos
                    print("Info : ", infoterm)
                    infoterm = infoterm +1
                    bioinfos.writerow([id, rl, nm, trad, birth, death, sal, \
                                       where, h, sp, books, tri, quotes])
                    
                if not (phy=='' and pher==''): # if there is biographies
                    print("Bio : ", bioterm)
                    bioterm = bioterm +1
                    biograhies.writerow([id, phy, pher])
                    
                id = id+1
                
            else: #the person is in [merge_people.csv]
                if not (rl=='' and nm=='' and birth=='' and h=='' and\
                       death=='' and sp=='' and tri=='' and books=='' and \
                        quotes=='' and sal=='' and  trad=='' and where==''):
                    print("Info : ", infoterm)
                    infoterm = infoterm +1
                    bioinfos.writerow([personid, rl, nm, trad, birth, death, sal, \
                                       where, h, sp, books, tri, quotes])
                if not (phy=='' and pher==''):
                    print("Bio : ", bioterm)
                    bioterm = bioterm +1
                    biograhies.writerow([personid, phy, pher])      
    
```

```python
add_bioinfos() 
```
